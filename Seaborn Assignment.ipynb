```python
#Q4. use dimonds data from seaborn plot histogram for the price column. use the hue parameter for the cut column of dia data
```

```python
#Ans 
sbn.load_dataset('diamonds')
```

```python
diamond=sbn.load_dataset('diamonds')
```

```python
sbn.displot(diamond['price'])
```

```python
sbn.relplot(x=diamond.color , y=diamond.price, data = diamond , hue='cut')
```

```python
#Q6. use the flights dataset from seaborn to plot a heatmap.
```
